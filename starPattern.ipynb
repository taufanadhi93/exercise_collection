```python
# creating stars pattern

stars = ''
size = 5

for i in range(size):
    for j in range(size):
        stars += '* '
    stars += '\n'
print(stars)

stars2 = ''
for i in range(size):
        for j in range(size-i):
            stars2 += '* '
        stars2 += '\n'
print(stars2)

print(stars+stars2)
```

```python
stars3 = ''
size = int(input('Masukan jumlah size yang diinginkan: '))
  
for i in range(size):
    for j in range(size-i):
        stars3 += '* '
    stars3 += '\n'

for i in range(1,size):
    for j in range(i+1):
        stars3 += '* '
    stars3 += '\n'
print(stars3)
```
