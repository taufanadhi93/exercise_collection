```python
import math

bb=int(input('Masukan berat badan anda: (kg)'))
tb=int(input('Masukan tinggi badan anda: (cm)'))

print('Massa tubuh anda adalah ', bb,' kg & Tinggi tubuh anda adalah ', tb/100, ' m')
imt=bb/(math.pow(tb/100,2))

if imt > 0 and imt < 18.5:
    print('IMT anda adalah ', imt,'Berat badan anda kurang')
elif imt >= 18.5 and imt <24.9:
    print('IMT anda adalah ', imt,'Berat badan anda ideal')
elif imt >= 25 and imt <29.9:
    print('IMT anda adalah ', imt,'Berat badan anda berlebih')
elif imt >= 30 and imt <39.9:
    print('IMT anda adalah ', imt,'Berat badan anda sangat berlebih')
elif imt >= 39.9:
    print('IMT anda adalah ', imt,'Anda OBESITAS')
```
