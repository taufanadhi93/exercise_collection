```python
ap=int(input('Masukan jumlah apel :'))
jr=int(input('Masukan jumlah jeruk :'))
ag=int(input('Masukan jumlah anggur :'))

totalAp=ap*10000
totalJr=jr*15000
totalAg=ag*20000
totalBelanja=totalAp+totalJr+totalAg
print('Detail belanja anda adalah', totalBelanja)
uangBelanja=int(input('Masukan jumlah uang anda :'))

if uangBelanja < totalBelanja:
    print('Transaksi anda dibatalkan \n Uang anda kurang', totalBelanja-uangBelanja)
elif uangBelanja == totalBelanja:
    print('Terima Kasih')
else :
    sisa=uangBelanja-totalBelanja
    print('Terima Kasih')
    print('Uang kembalian anda adalah',sisa)
```
