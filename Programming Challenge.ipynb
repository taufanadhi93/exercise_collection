```python
# prime number
a = []
answer = []

x = int(input("How many number you want to check ? "))
for i in range(x):
  b = int(input())
  a.append(b)

for j in range(len(a)):
    if a[j]> 1:
        if a[j]%2==0 or a[j]%3==0 or a[j]%5==0 or a[j]%7==0:
            answer.append('No')
        else:
            answer.append('Yes')
    else:
        answer.append('Yes')

for k in range(len(answer)):
    print(answer[k])
```
